# Task 1
## Write a program that allows you to enter 4 numbers and stores them in a file called “Numbers”

```python
numbers = open("Numbers.txt", "w+")
num_list = input("Enter numbers separated by space.").split()

for num in num_list:
    numbers.write(str(num) + "\n") #file handle gets to the end of file

numbers.seek(0) #takes the file handle to the begining
print(numbers.read())
numbers.close()



#or if we need it to be stricly 4 numbers:

# numbers = open("Numbers.txt", "w+")
# num_list = input("Enter 4 numbers separated by space.").split()

# while len(num_list) != 4:
#      num_list = input("Please enter precisely 4 numbers separated by space.").split()

# for num in num_list:
#     numbers.write(str(num) + "\n") #file handle gets to the end of file

# numbers.seek(0) #takes the file handle to the begining
# print(numbers.read())
# numbers.close()
```
